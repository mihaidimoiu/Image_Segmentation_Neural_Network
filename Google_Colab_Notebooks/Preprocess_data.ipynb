Load libraries to perform processing

```
import os
from numpy import asarray
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from numpy import savez_compressed
```

Load Google Drive API

```
from google.colab import drive
```

Mouting Google Drive for retriving and saving images

```
drive.mount('/content/drive/')
```

Path to Images location in Google Drive

```
CODE_PATH_DIR = "drive/My Drive/Disertatie/"
TRAIN_IMAGES_PATH = os.path.join(CODE_PATH_DIR, os.path.join("Train_Images", "Aerial/"))
OUTPUT_IMAGES_PATH = os.path.join(CODE_PATH_DIR, os.path.join("Output_Images", "Aerial/"))
```

Defining size for images

```
size = (256, 256)
```

Output filename

```
FILENAME = "map_" + str(size[0]) + "_" + str(size[1]) + ".npz"
FILENAME_SAVE_PATH = os.path.join(CODE_PATH_DIR, FILENAME)
print("Output filename: " + str(FILENAME))
print("Save path: " + str(FILENAME_SAVE_PATH))
```

Defining function for loading images from directory and returning an array

```
def LoadImages(path, size = (256, 256)):
    src_list = list()
    for filename in os.listdir(path):
        pixels = load_img(path + filename, target_size=size)
        pixels = img_to_array(pixels)
        src_list.append(pixels)
        
    return asarray(src_list)
```

Processing images

```
src_images = LoadImages(TRAIN_IMAGES_PATH, size)
print('Loaded: ', src_images.shape)

src_images2 = LoadImages(OUTPUT_IMAGES_PATH, size)
print('Loaded: ', src_images2.shape)
```

Saving array into Google Drive

```
savez_compressed(FILENAME_SAVE_PATH, src_images, src_images2)
print('Saved dataset: ', FILENAME_SAVE_PATH)
```

Umounting Google Drive

```
drive.flush_and_unmount()
```
